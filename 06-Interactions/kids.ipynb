```python
import pandas as pd

kids = pd.read_csv("https://github.com/febse/data/raw/main/econ/childiq.csv")

kids.head()
```

```python
import statsmodels.formula.api as smf

model = smf.ols("kid_score ~ mom_hs * mom_iq", data=kids)
results = model.fit()
print(results.summary())
```

```python
import seaborn as sns

sns.lmplot(x="mom_iq", y="kid_score", hue="mom_hs", data=kids)
```
